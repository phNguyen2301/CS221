```python
def preprocess(data):
  result = []
  for line in data:
    sentence = []
    line_split = line.split()
    for word_pos in line_split:
      w, pos = word_pos.split('/')
      sentence.append((w,pos))
    result.append(sentence)
  return result
```

```python
# compute Emission Probability
def word_given_tag(word, tag, train_bag = train_tagged_words):
  tag_list = [pair for pair in train_bag if pair[1]==tag]
  count_tag = len(tag_list)#total number of times the passed tag occurred in train_bag
  w_given_tag_list = [pair[0] for pair in tag_list if pair[0]==word]
  #now calculate the total number of times the passed word occurred as the passed tag.
  count_w_given_tag = len(w_given_tag_list)
  return (count_w_given_tag, count_tag)
```

```python
# compute Transition Probability
def t2_given_t1(t2, t1, train_bag = train_tagged_words):
  tags = [pair[1] for pair in train_bag]
  count_t1 = len([t for t in tags if t==t1])
  count_t2_t1 = 0
  for index in range(len(tags)-1):
      if tags[index]==t1 and tags[index+1] == t2:
          count_t2_t1 += 1
  return (count_t2_t1, count_t1)
```

```python
import numpy as np
```

```python
# creating t x t transition matrix of tags, t= no of tags
# Matrix(i, j) represents P(jth tag after the ith tag)
tags_matrix = np.zeros((len(tags), len(tags)), dtype='float32')
for i, t1 in enumerate(list(tags)):
    for j, t2 in enumerate(list(tags)): 
        tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/t2_given_t1(t2, t1)[1]
 
print(tags_matrix)
```

```python
import pandas as pd
```

```python
# convert the matrix to a df for better readability
#the table is same as the transition table shown in section 3 of article
tags_df = pd.DataFrame(tags_matrix, columns = list(tags), index=list(tags))
display(tags_df)
```

```python
def Viterbi(words, train_bag = train_tagged_words):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    for key, word in enumerate(words):
        p = [] 
        for tag in T:
            if key == 0:
                transition_p = -10000
            else:
                transition_p = tags_df.loc[state[-1], tag]
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
        pmax = max(p)
        state_max = T[p.index(pmax)] 
        state.append(state_max)
    return list(zip(words, state))
```

```python
test_tagged_words = [tup for sent in test_set for tup in sent]
test_untagged_words = [tup for sent in test_set for tup,_ in sent]
# test_untagged_words
tagged_seq = Viterbi(test_untagged_words)

# accuracy
check = [i for i, j in zip(test_tagged_words, tagged_seq) if i == j] 
 
accuracy = len(check)/len(tagged_seq)
print('Viterbi Algorithm Accuracy: ',accuracy*100)
```

```python
from collections import Counter, defaultdict
```

```python
def ngrams(text, n):
  Ngrams = []
  for i in range(len(text)): 
    Ngrams.append(tuple(text[i: i + n]))
  return Ngrams
```

```python
def bigram_counts(tags):
  for i_tag_bigram in ngrams(tags, 2):
      if i_tag_bigram in bigram_cnt:
          bigram_cnt[i_tag_bigram] += 1
      else:
          bigram_cnt[i_tag_bigram] = 1
  return bigram_cnt
```

```python
def unigram_counts(tags):
  for tag in tags:
      if tag in unigram_cnt:
          unigram_cnt[tag] += 1
      else:
          unigram_cnt[tag] = 1
  return unigram_cnt
```

```python
def tag_word_counts(tagged_words):
      for tag, word in tagged_words:
          tag_count[tag] += 1
          if (word, tag) in tag_word_count:
              tag_word_count[(tag, word)] += 1
          else:
              tag_word_count[(tag, word)] = 1
      return tag_word_count
```

```python
def transition_probabilty(tags):
  bigrams = ngrams(tags, 2)
  for bigram in bigrams:
      transition_probabilities[bigram] = bigram_cnt[bigram] / unigram_cnt[bigram[0]]
  return transition_probabilities
```

```python
def emmission_probabilty(tagged_words):
  for tag, word in tagged_words:
      emmission_probabilities[tag, word] = tag_word_count[tag, word] / tag_count[tag]
  return emmission_probabilities
```

```python
def initial_probabilities(tag):
  return transition_probabilities["<s>", tag]
```
