```python
import matplotlib.pyplot as plt
```

```python
dic = {"one_gram": len(one_gram), "bi_grams": len(bi_grams), "tri_grams": len(tri_grams)}
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        print(i, y[i])
        plt.text(i,y[i]+0.01*y[i],y[i])
```

```python
len(dic.keys())
```

```python
fig = plt.figure(figsize = (10, 5))

plt.bar(dic.keys(), dic.values(), color ='blue',
        width = 0.4)
addlabels(dic.keys(), list(dic.values()))
plt.xlabel("Vocab")
plt.ylabel("Amount")
plt.title("Number of Vocab")
plt.show()
```
