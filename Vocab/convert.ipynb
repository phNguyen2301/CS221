```python
from pysenal import write_lines
import string
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re
from itertools import chain
PAD_WORD = '<pad>'
UNK_WORD = '<unk>'
BOS_WORD = '<s>'
EOS_WORD = '</s>'
DIGIT_WORD = '<digit>'
SEP_WORD = '<sep>'
vocab_size = 50000
num_and_punc_regex = re.compile(r'[\-—<>{,(?\\.\'%]|\d+([.]\d+)?', re.IGNORECASE)
num_regex = re.compile(r'\d+([.]\d+)?')
stemmer = PorterStemmer()
```

```python
def tokenize(text, is_lower, is_stem):
    text = num_and_punc_regex.sub(r' \g<0> ', text)
    tokens = word_tokenize(text)
    if is_lower:
        tokens = [token.lower() for token in tokens]
    if is_stem:
        tokens = [stemmer.stem(token) for token in tokens]
    for idx, token in enumerate(tokens):
        token = tokens[idx]
        if num_regex.fullmatch(token):
            tokens[idx] = DIGIT_WORD
    return tokens
```

```python
test = tokenize(tokens[1], True, True)
print(test)
```

```python
def build_vocab(tokens):
    vocab = [PAD_WORD, UNK_WORD, BOS_WORD, EOS_WORD, DIGIT_WORD, SEP_WORD]
    vocab.extend(list(string.digits))
    token_counter = Counter(tokens).most_common(vocab_size)
    for token, count in token_counter:
        vocab.append(token)
        if len(vocab) >= vocab_size:
            break
    return vocab
```

```python
tokens_temp = []
for text in tokens:
  temp = tokenize(text, True, True)
  tokens_temp.append(temp)
tokens_final = list(chain.from_iterable(tokens_temp)) 
vocab = build_vocab(tokens_final)
write_lines("vocab.txt", vocab)
```
