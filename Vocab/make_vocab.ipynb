```python
import string
from collections import Counter
import re
from itertools import chain

PAD_WORD = '<pad>'
UNK_WORD = '<unk>'
BOS_WORD = '<s>'
EOS_WORD = '</s>'
DIGIT_WORD = '<digit>'
SEP_WORD = '<sep>'
vocab_size = 50000

num_and_punc_regex = re.compile(r'[\-—<>{,(?\\.\'%]|\d+([.]\d+)?', re.IGNORECASE)
num_regex = re.compile(r'\d+([.]\d+)?')
```

```python
def tokenize(text, is_lower=True):
  text = num_and_punc_regex.sub(r' \g<0>', text)
  tokens = text.split()
  if is_lower:
        tokens = [token.lower() for token in tokens]
  for idx, token in enumerate(tokens):
      token = tokens[idx]
      if num_regex.fullmatch(token):
          tokens[idx] = DIGIT_WORD
  return tokens
```

```python
def build_vocab(tokens):
    vocab = [PAD_WORD, UNK_WORD, BOS_WORD, EOS_WORD, DIGIT_WORD, SEP_WORD]
    vocab.extend(list(string.digits))
    # token_counter = Counter(tokens).most_common(vocab_size)
    token_counter = Counter(tokens).most_common()
    for token, count in token_counter:
        vocab.append(token)
        if len(vocab) >= vocab_size:
            break
    return vocab
```

```python
tokens_temp = []
for text in tokens:
  temp = tokenize(text)
  tokens_temp.append(temp)
tokens_final = list(chain.from_iterable(tokens_temp))
```

```python
vocab = build_vocab(tokens_final)
```

```python
with open("vocab.txt", "w") as txt_file:
    for line in vocab:
        txt_file.write("".join(line) + "\n")
```
