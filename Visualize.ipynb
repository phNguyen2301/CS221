```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open('/content/gold.txt')
listContent = f.readlines()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(listContent, test_size=0.2, random_state=1, shuffle=True)
print(len(train),train)
print(len(test),test)
```

```python


total = 0
maxWord = 0
sentenceMax = ''
minWord = 1000000000
sentenceMin = ''
max = 0
min = 1000000000
print(len(listContent))
for line in listContent:
  totalWord = line.count(' ')
  total += totalWord
  if(totalWord > maxWord):
    maxWord = line.count('/')
    sentenceMax = line
  if(totalWord < minWord):
    minWord = totalWord
    sentenceMin = line
  if(totalWord > max):
    max = len(line)
  if(totalWord < min):
    min = len(line)
dic = {
    
}

for line in listContent:
  words = line.split()
  for word in words:
    tem = word.split('/')
    if(tem[1] in dic.keys()):
      dic[tem[1]] += 1
    else:
      dic[tem[1]] = 0


dic = dict(sorted(dic.items(), key=lambda x: x[1])) 

print('maxWord', maxWord, sentenceMax)
print('minWord', minWord, sentenceMin)
print('average', round(total/len(listContent)))
print(max, min)
print('dic', dic)
print('total label', sum(dic.values()))

fig = plt.figure(figsize = (10, 5))

plt.bar(dic.keys(), dic.values(), color ='maroon',
        width = 0.4)
 
plt.xlabel("Label")
plt.show()
```
